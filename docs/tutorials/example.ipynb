### Subsection B

```python
plt.figure(figsize=(6.0,3))
t_save = np.linspace(0, 10, 100)
plt.plot(t_save, np.cos(2 * t_save), label=r'Ground')
plt.plot(t_save, np.sin(2 * t_save), label=r'Excited')

# formatting
plt.xlabel('Time')
plt.ylabel('Function')
plt.xlim(0, 10)
plt.ylim(-1, 1)
plt.legend(frameon=True)
plt.style.use('bmh')

# display plot
plt.show()
```
